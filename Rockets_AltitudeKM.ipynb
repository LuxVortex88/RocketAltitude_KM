```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Mock Rockets name
Rockets = ['Raptor9', 'Savannah', 'Lunax', 'Sxer',
]

# Sample Rockets Altitude in KM
add_value_altitude_km = [2000,5000,7000,8000,]
count = [30, 21, 76, 25]

print(Rockets)
print(add_value_altitude_km)

plt.plot(add_value_altitude_km)
plt.xlabel('rockets')
plt.ylabel('altitude_km')
plt.title('rocket vs altitude_km')
plt.show()

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
for i in range(len(Rockets)):
    ax.bar(Rockets[i], add_value_altitude_km[i], count[i])
    ax.set_xlabel('rockets')
    ax.set_ylabel('altitude_km')
    ax.set_zlabel('count')
    ax.set_title('rocket and altitude_km')
```
